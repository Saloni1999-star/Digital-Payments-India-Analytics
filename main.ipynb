```python
# Step 1: Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Step 2: Create the dataset manually
data = {
    'Year': ['2020-21', '2021-22', '2022-23', '2023-24', '2024-25'],
    'Volume_Cr': [4370.68, 7197.68, 11393.82, 16443.02, 18120.82],
    'Value_LCr': [1414.58, 1744.01, 2086.85, 2428.24, 2330.72]
}

# Step 3: Convert to DataFrame
df = pd.DataFrame(data)

# Step 4: View the DataFrame
print(df)
```

```python
# Set the figure size and style
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Plot Volume and Value lines
sns.lineplot(x='Year', y='Volume_Cr', data=df, marker='o', label='Transaction Volume (Cr)', linewidth=2)
sns.lineplot(x='Year', y='Value_LCr', data=df, marker='o', label='Transaction Value (Lakh Cr)', linewidth=2)

# Add titles and labels
plt.title('📊 Digital Payment Trends in India (2020–2025)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculate % Growth for Volume and Value
df['Volume_YoY_%'] = df['Volume_Cr'].pct_change().round(2) * 100
df['Value_YoY_%'] = df['Value_LCr'].pct_change().round(2) * 100

# Display the updated DataFrame
print(df)
```

```python
# Line chart: Transaction Volume & Value
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

sns.lineplot(x='Year', y='Volume_Cr', data=df, marker='o', label='Volume (Cr)', linewidth=2)
sns.lineplot(x='Year', y='Value_LCr', data=df, marker='o', label='Value (Lakh Cr)', linewidth=2)

plt.title('Digital Payments Volume vs. Value (2020–2025)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.tight_layout()
plt.savefig("digital_payments_trend.png")  # Save as PNG
plt.show()
```

```python
# Bar chart: YoY % Growth
plt.figure(figsize=(10, 6))
bar_width = 0.35
x = range(len(df))

plt.bar(x, df['Volume_YoY_%'], width=bar_width, label='Volume YoY %', color='skyblue')
plt.bar([i + bar_width for i in x], df['Value_YoY_%'], width=bar_width, label='Value YoY %', color='salmon')

plt.xticks([i + bar_width / 2 for i in x], df['Year'])
plt.ylabel('Growth %')
plt.title('Year-on-Year Growth: Volume vs Value (%)')
plt.legend()
plt.tight_layout()
plt.savefig("digital_payments_yoy_growth.png")  # Save as PNG
plt.show()
```
